```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
```

```python
train_df = pd.read_csv("/kaggle/input/week4-practice-hackathon-2024/Train.csv")
```

```python
def normalize_num_cols(df):
    scaler = StandardScaler()
    numeric_cols = df.select_dtypes(include=['number']).columns
    
    df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
    return df
```

```python
X = train_df.drop(['class'], axis=1)
y = train_df['class']
```

```python
X.dtypes
```

```python
# Identify columns with any null values
columns_with_null = X.columns[X.isnull().any()].tolist()

print("Columns with null values:", columns_with_null)
```

```python
normalize_num_cols(X)
X.head()
```

```python
y.head()
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.3)
```

```python
print(y_train.value_counts())
```

```python
from sklearn.naive_bayes import GaussianNB
lr=GaussianNB()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_val)
print(y_pred)
```

```python
f1 = f1_score(y_val, y_pred)
print(f"F1 score: {f1}")
```
